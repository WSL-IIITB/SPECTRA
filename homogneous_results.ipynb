```python
def df_prop(dict, prop):
    dupDict = copy.deepcopy(dict)
    list_keys = list(dict.keys())
    # print(list_keys)
    for i in list_keys:
        # print(dupDict,i)
        for j in dupDict[i]:
            dupDict[i][j] = np.mean(getNetworkProp(dupDict[i][j].getGraph(), prop))
    print(dupDict)
    df = pd.DataFrame.from_dict(dupDict)
    return df
```
